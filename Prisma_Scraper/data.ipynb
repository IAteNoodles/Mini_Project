```python
def clean_text(text):
    # Remove newline and tab characters
    text = text.replace('\n', ' ').replace('\t', ' ')
    # Remove extra spaces
    text = ' '.join(text.split())
    return text
```

```python
def remove_urls(text):
    import re
    url_pattern = r'http\S+|www\S+'
    return re.sub(url_pattern, '', text)
```

```python
def remove_punctuation(text):
    import string
    return text.translate(str.maketrans('', '', string.punctuation))
```

```python
def remove_emojis(text):
    import re
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        "]+",
        flags=re.UNICODE,
    )
    return emoji_pattern.sub(r'', text)
```

```python
pipiline = [clean_text, remove_urls, remove_punctuation, remove_emojis]
```

```python
for func in pipiline:
    data_fixed['article'] = data_fixed['article'].apply(func)
```

```python
from transformers import AutoTokenizer
```

```python
model_id = "distilbert-base-cased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
word_embeddings = tokenizer(data_fixed['article'].tolist(), padding=True, truncation=True, return_tensors="pt")
```

```python
# Use K Means to cluster the articles based on their embeddings
from sklearn.cluster import KMeans
# Use Elbow method or Silhouette score to determine optimal number of clusters
import numpy as np
inertias = []
for k in range(1, 11):
    kmeans_temp = KMeans(n_clusters=k, random_state=42)
    kmeans_temp.fit(word_embeddings['input_ids'])
    inertias.append(kmeans_temp.inertia_)

plt.plot(range(1, 11), inertias, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()

# Based on the elbow plot, set the optimal number of clusters
num_clusters = 4

kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(word_embeddings['input_ids'])

```

```python
# See the graph of the clustersimport matplotlib.pyplot as plt
from sklearn.decomposition import PCA
# Reduce dimensions for visualization
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(word_embeddings['input_ids'])
plt.figure(figsize=(10, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('K-Means Clustering of Articles')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
